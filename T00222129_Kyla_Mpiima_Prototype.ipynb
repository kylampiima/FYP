```python
from gtts import gTTS

# Text to be converted to audio
#text = "I have a sp ee ee eech impediment."

stuttered_sentences = [
    "Sa-sa-sometimes I fe-ee-eel a lii-little shy.",
    "To-to-today is a gray-great day.",
    "I-I-I really love to play the guitar.",
    "Wait for me a-a-at the corner of the street.",
    "Thank u-you for your help today.",
    "Sa-sa-sometimes I just want to relax and read.",
    "Coo-coo-could you please pass the salt?",
    "He-he-here comes the sun after the rain.",
    "May-may-maybe we can go to the movies later.",
    "Fray-fray-friends are important in our lives.",
    "u-u-you should try that new restaurant downtown.",
    "Lii-lii-little things can make a big difference."
]

# Loop through each sentence and create a gTTS object
for i, sentence in enumerate(stuttered_sentences):
    # Create a gTTS object
    tts = gTTS(text=sentence, lang='en', slow=True)

    # Save the audio file with a unique name
    filename = f"speech_impediment_gtts_{i+1}.wav"
    tts.save(filename)
    print(f"Saved: {filename}")


# Create a gTTS object
#tts = gTTS(text=text, lang='en')

#slow=True

# Save the audio file
#tts.save("speech_impediment_gtts.wav")
```

```python
import soundfile as sf
import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model =Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-960h")
```

```python
input_audio,_ = librosa.load(file_name,sr=16000)
```

```python
input_values = tokenizer(input_audio,return_tensors="pt").input_values
logits = model(input_values).logits
predicted_ids = torch.argmax(logits,dim=-1)
text = tokenizer.batch_decode(predicted_ids)[0]
```
